```python
import os

input_bert_dict_file = \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/BERT_merge_file/data/PubMed-VLSP.txt"
total_corpus_file = \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/BERT_merge_file/BERT_create_corpus/view_data/data/total_corpus.txt"

def create_bert_dict(input_bert_dict_file, total_corpus_file):
    input_bert_dict_ls = []

    with open(os.path.join(input_bert_dict_file), 'r') as in_file:
        lines = in_file.readlines()
        for line in lines:
            input_bert_dict_ls.append(line.split(" ")[0])

    total_corpus_ls = []
    with open(os.path.join(total_corpus_file), 'r') as in_file:
        line = in_file.readlines()[0]
        for l in line.split("\t"):
            total_corpus_ls.append(l)

    print("input_bert_dict_ls ", len(input_bert_dict_ls), input_bert_dict_ls)
    print("total_corpus_ls ", len(total_corpus_ls), total_corpus_ls)

    return input_bert_dict_ls, total_corpus_ls
input_bert_dict_ls, total_corpus_ls = create_bert_dict(input_bert_dict_file, total_corpus_file)

```
