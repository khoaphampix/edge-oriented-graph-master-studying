```python
import json

def extract_quoted_words(input_file_path, output_file_path):
    data = {}
    with open(input_file_path, 'r') as f:
        for line in f:
            line_parts = line.strip().split('\t')
            code = line_parts[0]
            words = {}
            for part in line_parts:
                if '"' in part:
                    word = part.strip('"')
                    words[f'"{word}'] = word
            if words != {}:
                data[code] = words


    with open(output_file_path, 'w') as f:
        json.dump(data, f)

out_file_vlsp_raw_double_quote= \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/Re_check_data/error_double_quote_data/merge_all_line_2_find_double_quote"

out_json_file_vlsp_raw_double_quote = \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/Re_check_data/error_double_quote_data/json_find_double_quote.json"


extract_quoted_words(out_file_vlsp_raw_double_quote, out_json_file_vlsp_raw_double_quote)
```

```python
import os
import json

def replace_strings(folder_path, mapping_file):
    # Read mapping file as a dictionary
    with open(mapping_file, 'r') as f:
        mapping = json.load(f)
        
    # Loop through subfolders in the main folder
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            # Check if subfolder name is a key in the mapping dictionary
            if subfolder in mapping:
                # Read the CURATION_USER.tsv file
                file_path = os.path.join(subfolder_path, "CURATION_USER.tsv")
                with open(file_path, 'r') as f:
                    content = f.read()
                
                # Replace strings based on the mapping dictionary
                for key, val in mapping[subfolder].items():
                    content = content.replace(key, val)
                
                # print(content)
                # print(file_path[-40:])
                # Write the updated file
                with open(file_path, 'w') as f:
                    f.write(content)
            # else:
            #     print(f"Skipping folder {subfolder} because it's not in the mapping dictionary")

vlsp_folder_path = \
        "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/VLSP2020_RE_dev"
    # "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/double_quote_file_check/double_quote_files"
mapping_file = \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/Re_check_data/error_double_quote_data/json_find_double_quote.json"
replace_strings(vlsp_folder_path, mapping_file)
```
