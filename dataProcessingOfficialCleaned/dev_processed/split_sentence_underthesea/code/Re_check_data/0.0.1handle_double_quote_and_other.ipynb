```python
import os
error_on_files = \
[
    "23351888.conll", "23354946.conll", "23353861.conll", "23357135.conll", "23354944.conll", "23357329.conll", "23354202.conll",
    "23353945.conll", "23356638.conll", "23353867.conll", "23357336.conll", "23354285.conll", "23351647.conll", "23352755.conll", 
    "23351809.conll", "23356765.conll", "23353825.conll", "23352825.conll", "23355773.conll", "23351956.conll", "23352752.conll", 
    "23351424.conll", "23355470.conll", "23356093.conll", "23352886.conll", "23351969.conll", "23356767.conll", "23354085.conll", 
    "23354648.conll", "23351627.conll", "23352708.conll", "23351316.conll", "23351931.conll", "23352754.conll", "23353794.conll", 
    "23356745.conll", "23354920.conll", "23351391.conll", "23357308.conll", "23351519.conll", "23354793.conll", "23351856.conll", 
    "23351435.conll", "23352743.conll", "23351651.conll", "23351987.conll", "23356907.conll", "23366722.conll", "23354400.conll", 
    "23353857.conll", "23351437.conll", "23352704.conll", "23357263.conll", "23351981.conll", "23351307.conll", "23353830.conll", 
    "23352872.conll", "23354912.conll", "23353780.conll", "23355228.conll", "23351615.conll", "23353931.conll", "23351946.conll", 
    "23355132.conll", "23353950.conll", "23353954.conll", "23356771.conll", "23353727.conll", "23355626.conll", "23351571.conll", 
    "23356494.conll", "23354953.conll", "23352892.conll", "23351493.conll", "23351959.conll", "23353913.conll", "23351416.conll"
]

def find_words_in_files(folder_path):
    """
    Reads all files in the specified folder and its subfolders,
    finds occurrences of "\t" and extracts the word between "\t"
    and "\t", then stores all words in a dictionary with the key
    being the subfolder name.
    """
    words_dict = {}
    for dirpath, dirnames, filenames in os.walk(folder_path):
        subfolder_name = os.path.basename(dirpath)
        words = set()
        # if subfolder_name not in error_on_files:
        #     continue
        print(subfolder_name)

        for filename in filenames:
            if filename not in ["CURATION_USER.tsv"]:
                continue
            print(filename)
            # get the full path of the file
            file_path = os.path.join(dirpath, filename)
            
            # read the file and find the words
            with open(file_path, 'r') as f:
                for line in f:
                    # print(line)
                    while True:
                        start_idx = line.find('\t"')
                        if start_idx == -1:
                            break
                        end_idx = line.find('\t', start_idx + 2)
                        if end_idx == -1:
                            break
                        word = line[start_idx + 2:end_idx]
                        words.add(word)
        if words:
            words_dict[subfolder_name] = words
    return words_dict
            


test_folder = \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/double_quote_file_check/double_quote_files"
words_dict = find_words_in_files(test_folder)
print(words_dict)

# # example usage:
# replace_word_in_files(test_folder)

```
