```python
import os
import re

def extract_strings_from_files(folder_path, output_file):
    with open(output_file, 'w') as f:
        for file_name in os.listdir(folder_path):
            # extract the code from the file name
            code = os.path.splitext(file_name)[0]

            # read the second line of the file
            with open(os.path.join(folder_path, file_name)) as file:
                lines = file.readlines()
                if len(lines) > 1:
                    second_line = lines[1]
                else:
                    second_line = ''

            # extract the matching strings from the second line using regex
            regex = r'\b([^\sA-Z]+)\.\s([A-Z][^\s]+)\b'
            matches = re.findall(regex, second_line)

            # write the results to the output file
            for match in matches:
                # print(match)
                f.write(f"{code}\t{match[0]}. {match[1]}\n")

folder_path_process = \
"/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/docs/common_info_embedd_files/split_passage_for_final_processed"

output_file_sent = \
"/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/clean_vietnamese_text/fix_split_sentence/view_data/data/wrong_split_sentence.txt"

extract_strings_from_files(folder_path_process, output_file_sent)

```
