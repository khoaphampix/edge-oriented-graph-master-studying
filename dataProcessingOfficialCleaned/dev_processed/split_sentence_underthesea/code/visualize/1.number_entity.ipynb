```python
import os

def count_words_in_subfolders(folder_path, out_file_path):
    # words_to_count = ['LOCATED', 'AFFILIATION', 'PART - WHOLE', 'PERSONAL - SOCIAL']
    words_to_count = ['LOCATED', 'AFFILIATION', 'WHOLE', 'PERSONAL - SOCIAL']
    counts = {}
    print(os.listdir(folder_path))
    for subfolder_name in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder_name)
        # print(subfolder_path[-40:])
        if os.path.isdir(subfolder_path):
            # print(subfolder_path)
            file_path = os.path.join(subfolder_path, 'CURATION_USER.tsv')
            if os.path.isfile(file_path):
                # print(file_path)

            
                with open(file_path, 'r') as subfolder_file:
                    file_contents = subfolder_file.read()
                    word_counts = [file_contents.lower().count(word.lower()) for word in words_to_count]
                    total_count = sum(word_counts)
                    counts[subfolder_name] = total_count
    with open(out_file_path, 'w') as f:
        for code, count in counts.items():
            f.write(f"{code}: {count}\n")




train = \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/VLSP2020_RE_train"


dev = \
    "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/VLSP2020_RE_dev"

dev_out_file_raw = "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/visualize/data/total_relation/dev"


train_out_file_raw = "/Users/n2t2k/Documents/Studying/Master/Thesis/InProgress/Coding/ORIGIN_RUN_ALL_edge-oriented-graph-master-studying/dataProcessingOfficialCleaned/dev_processed/split_sentence_underthesea/code/visualize/data/total_relation/train"

count_words_in_subfolders(dev, dev_out_file_raw)
count_words_in_subfolders(train, train_out_file_raw)


```
